```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('punkt_tab')
sentence='i woke up at 7 am'
tokens=word_tokenize(sentence)
print(tokens)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
sentence='she was going to a movie. she regulary goes to movies as she loves them. she is an optimitic person'
tokens=word_tokenize(sentence)
stop_words=set(stopwords.words('english'))
filtered_tokens=[word for word in tokens if word.lower() not in stop_words]
print(filtered_tokens)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
nltk.download('punkt')
nltk.download('punkt_tab')
sentence='The children were playing in the garden happily, enjoying the sunshine. They danced, laughed, jumped, and shouted with excitement, forgetting their worries and focusing only on the moment of joy'
tokens=word_tokenize(sentence)
stemmer=PorterStemmer()
stemmed_tokens=[stemmer.stem(word) for word in tokens]
print(stemmed_tokens)


```
